# Feature Engineering 

```python

# find categorical variables

categorical = [var for var in africa_systemic_crisis.columns if africa_systemic_crisis[var].dtype=='O']

print('There are {} categorical variables'.format(len(categorical)))
```

```python
africa_systemic_crisis[categorical].head()
```

```python
# find categorical variables

numerical = [var for var in africa_systemic_crisis.columns if africa_systemic_crisis[var].dtype!='O']

print('There are {} categorical variables'.format(len(numerical)))
```

```python
africa_systemic_crisis[numerical].head()
```

```python
year_vars = [var for var in numerical if 'year' in var]

year_vars
```

```python
africa_systemic_crisis[year_vars].head()
```

```python
# let's visualise the values of the discrete variables
discrete = []

for var in numerical:
    if len(africa_systemic_crisis[var].unique()) < 20 and var not in year_vars:
        print(var, ' values: ', africa_systemic_crisis[var].unique())
        discrete.append(var)
print()
print('There are {} discrete variables'.format(len(discrete)))
```

```python
# find continuous variables
# let's remember to skip the Id variable and the target variable SalePrice
# which are both also numerical

continuos = [var for var in numerical if var not in discrete and var not in year_vars]

print('There are {} numerical and continuous variables'.format(len(continuos)))
```

```python
africa_systemic_crisis[continuos].head()
```

Since there is no missing values, we proceed to check for outliers

```python
# let's make boxplots to visualise outliers in the continuous variables 
# and histograms to get an idea of the distribution

for var in continuos:
    plt.figure(figsize=(15,4))
    plt.subplot(1, 2, 1)
    fig = africa_systemic_crisis.boxplot(column=var)
    fig.set_title('')
    fig.set_ylabel(var)
    
    plt.subplot(1, 2, 2)
    fig =africa_systemic_crisis[var].hist(bins=20)
    fig.set_ylabel('')
    fig.set_xlabel(var)

    plt.show()
```

```python
for var in discrete:
    (africa_systemic_crisis.groupby(var)[var].count() / np.float(len(africa_systemic_crisis))).plot.bar()
    plt.ylabel('Percentage of observations per label')
    plt.title(var)
    plt.show()
```

```python
for var in discrete:
    africa_systemic_crisis.groupby(var)['currency_crises'].median().plot()
    plt.ylabel('Median house Price per label')
    plt.title(var)
    plt.show()
```

```python
banking_crisis = pd.get_dummies(africa_systemic_crisis['banking_crisis'],drop_first=True)
banking_crisis.head()
```

```python
africa_systemic_crisis.drop(['cc3','country','banking_crisis'],axis=1,inplace=True)
```

```python
africa_systemic_crisis = pd.concat([africa_systemic_crisis,banking_crisis],axis=1)
africa_systemic_crisis.head()
```

```python
africa_systemic_crisis.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = africa_systemic_crisis.drop(['no_crisis'], axis=1)
y = africa_systemic_crisis['no_crisis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=10)
X_train.shape, X_test.shape
```

# logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_regression = LogisticRegression()
logistic_regression.fit(X_train,y_train)
```

```python
predictions = logistic_regression.predict(X_test)
```

# Evaluating the model

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(classification_report(y_test,rfc_pred))
```

```python
print(confusion_matrix(y_test,rfc_pred))
```
